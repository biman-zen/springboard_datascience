```python
# Calculate the price change between on the following years
year_start = 1998
year_end = 2018
# Empty Lists
mean_start = []
mean_end = []

for index, row in df.iterrows():
    # Get the average homes of a given year
    mean_start.append(df.loc[index,year_list(year_start)].mean())
    mean_end.append(df.loc[index,year_list(year_end)].mean())

# Create an array for Price Ratio
price_ratio = np.array(mean_start)/np.array(mean_end)
# Create an array of % Price Change
percent_change = 100*(np.array(mean_end)-np.array(mean_start))/np.array(mean_start)

# Add new columns to the dataframe given the arrays calculated
df['PriceRatio'] = price_ratio
df['PercentChange'] = percent_change
df.head()
```

```python
# Create a new data frame with the borough names and calculation columns
borough_df = df[['Name','PriceRatio','PercentChange']].copy()
# Drop the other columns
borough_df = borough_df.iloc[0:33]
# Sort the dataframe from lowest to highest price ratio
sorted_borough_df = borough_df.sort_values(by='PriceRatio')
print('\nThe boroughs with the highest price change:\n')
print(sorted_borough_df.head())
print('\nThe boroughs with the lowest price change:\n')
print(sorted_borough_df.tail())
# Note that the lowest price ratio means that the prices have increased more. The percent change
```

```python
# Plot the new data frame
fig, ax = plt.subplots()
sorted_borough_df.plot(x='Name',y='PriceRatio',kind='bar',xlabel='',ylabel='PriceRatio',ax=ax,legend=False)
#help(ax.legend([]))
ax.set_title('Price Ratio of Average Home Prices by London Borough \nfrom 1998 to 20018')
# Plot a bar chart of the price ratio
#plt.bar(df.iloc[0:33,0],df.iloc[0:32,360])
# Rotate boroughs
plt.xticks(rotation=90)
plt.show()
```
